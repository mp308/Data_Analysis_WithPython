```python
#เลือกอันที่ไม่ว่าง
select_column = ['id','name','zipcode']
condition = w_restaurant['zipcode'].notna()
w_restaurant_selection = w_restaurant.loc[condition,select_column]
print(w_restaurant_selection.shape)
w_restaurant_selection.head(10)
```

```python
#4.4
new_w_restaurant = w_restaurant['zipcode'].dropna()

condition = new_w_restaurant['zipcode'].str.contains('60')
w_restaurant_selection = w_restaurant.loc[condition,:]
print(w_restaurant_selection.shape)
w_restaurant_selection.head(10)
```

```python
#การเชื่อมเงื่อนไข and or not
condition1 =  w_restaurant['zipcode'].notna()
condition2 =  w_restaurant['zipcode'].str.contains('60')
condition = condition1 & condition2
w_restaurant_selection = w_restaurant.loc[condition,:]
print(w_restaurant_selection.shape)
w_restaurant_selection.head(10)
```

```python
#ร้านใน เชียงใหม่ และ กาญ
condition1 = w_restaurant['city_name'] == 'Chiang Mai'
condition2 = w_restaurant['city_name'] == 'Kanchanaburi'
condition = condition1 | condition2
w_restaurant_selection = w_restaurant.loc[condition,:]
print(w_restaurant_selection.shape)
w_restaurant_selection.head(10)
```

```python
#rateting 2-4 and japan
condition1 = w_restaurant['weighted_average_rating'] >= 2
condition2 = w_restaurant['weighted_average_rating'] <= 4
condition3 = w_restaurant['category_name'] == 'Japanese'
condition = (condition1 & condition2) | condition3
w_restaurant_selection = w_restaurant.loc[condition,:]
print(w_restaurant_selection.shape)
w_restaurant_selection.head(10)
```

```python
#5.1
select_column = ['id','name','city_name','weighted_average_rating']
condition1 = (w_restaurant['weighted_average_rating'] >= 2) & (w_restaurant['city_name'] == 'Chiang Mai')
condition2 = (w_restaurant['weighted_average_rating'] >= 3) & (w_restaurant['city_name'] == 'Ratchaburi')
condition = condition1 | condition2
w_restaurant_selection = w_restaurant.loc[condition,select_column]
print(w_restaurant_selection.shape)
w_restaurant_selection.head(10)
```

```python
#5.2
select_column = ['id','name','avg_price','category_name']
condition1 = (w_restaurant['avg_price'] >= 200) & (w_restaurant['avg_price'] <= 200)
condition2 = w_restaurant['category_name'] == 'Japanese'
condition = condition1 | condition2
w_restaurant_selection = w_restaurant.loc[condition,select_column]
print(w_restaurant_selection.shape)
w_restaurant_selection.head(10)
```
