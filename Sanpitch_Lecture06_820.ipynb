```python
#plot barchart ของ ร้านใน zip code 20000, 20130 , 20150, 20170, 20190
#ที่มี reviews > 0
w_restaurant = pd.read_csv('w_restaurant.csv')
from matplotlib import pyplot as plt
plot_column = 'zipcode'
condition1 = w_restaurant['zipcode'] == '20000'
condition2 = w_restaurant['zipcode'] == '20130'
condition3 = w_restaurant['zipcode'] == '20150'
condition4 = w_restaurant['zipcode'] == '20170'
condition5 = w_restaurant['zipcode'] == '20190'
condition6 = w_restaurant['number_of_reviews'] > 0
condition = (condition1 | condition2 | condition3 | condition4 | condition5) & condition6
#.dropna() : ใช้ในการตัดข้อมูลที่เป็นค่าว่างทิ้ง
w_restaurant_selection = w_restaurant.loc[condition,plot_column].dropna()

#แปลง Series เป็น list
input_data = list(w_restaurant_selection)
#นับค่าใน list ด้วย Counter
output_data = Counter(input_data)

#plot bar chart
x = output_data.keys() #
y = output_data.values() #
#เปลี่ยนขนาด plot กว้าง 20 สูง 5
plt.rcParams['figure.figsize'] = [10, 5]
plt.bar(x, y)
```
