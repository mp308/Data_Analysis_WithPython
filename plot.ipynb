```python
#https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html
from matplotlib import pyplot as plt
x = [2549, 2550, 2551, 2552, 2553, 2554, 2555, 2556]
y = [3187.25, 3510.65, 3979.67, 5349.99, 6270.13, 6952.13, 7896.70, 8260.66]
# create a line chart, years on x-axis, gdp on y-axis
plt.plot(x, y, color='red', marker='x', linestyle='dashed')
# add a title
plt.title("Soap Exports by Year",color='blue')
# add a label to the y-axis
plt.ylabel("Millions of Baht")
plt.xlabel("Years")
plt.show()
```

```python
from matplotlib import pyplot as plt
movies = ["Lumber", "Veneer planks", "plywood", "Fibreboard", "Wooden photo frame"]
num_oscars = [21878.04, 67.10, 9117.83, 12116.23	, 1784.43]
# plot bars with left x-coordinates [0, 1, 2, 3, 4], heights [num_oscars]
plt.bar(range(len(movies)), num_oscars)
plt.title("Wood and wood products in 2554") # add a title
plt.ylabel("# of Academy Awards") # label the y-axis
# label x-axis with movie names at bar centers
plt.xticks(range(len(movies)), movies)
plt.show()
```

```python
#https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html
from matplotlib import pyplot as plt
x = [2550, 2551, 2552, 2553, 2554, 2555, 2556, 2557]
y1 = [21227.76	, 21814.38	, 20833.61, 20482.46, 17100.93, 20305.63, 16033.50	, 21085.91] #ปูน
y2 = [25351.83, 47809.19, 53312.63, 2814.04, 2923.65	, 2809.90, 2813.76, 2563.03] #หนังสือพิมพ์
y3 = [15499.89, 16390.40, 16600.07, 24789.19, 18288.42, 15015.89, 15043.02, 17526.17] #กระจก
# create a line chart, years on x-axis, gdp on y-axis
plt.plot(x, y1, color='red', marker='x', linestyle='dashed', label='cement')
plt.plot(x, y2, color='blue', marker='o', linestyle='solid', label='news paper')
plt.plot(x, y3, color='black', marker='o', linestyle='solid', label='glass')
plt.legend(loc=9)
# add a title
plt.title("Thai Product Exports by Year",color='blue')
# add a label to the y-axis
plt.ylabel("Millions of Baht")
plt.xlabel("Years")
plt.show()
```
